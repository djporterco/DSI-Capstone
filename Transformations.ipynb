```python
import pandas as pd
from matplotlib import pyplot
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('./analysis/BTC.csv')
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df = df.rename(index=str, columns={"votes_disliked": "dislike",
                              "votes_important": "important",
                              "votes_liked": "liked",
                              "votes_lol": "lol",
                              "votes_negative": "bearish",
                              "votes_positive": "bullish",
                              "votes_saved": "saved",
                              "votes_toxic": "toxic"})
```

```python
df['user_sent_opinions'] = df['bearish'] + df['bullish']

df['bullishness'] = ((df['bullish'] / df['user_sent_opinions']) * 100)

df['bearishness'] = -((df['bearish'] / df['user_sent_opinions']) * 100)
```

### Bullish Transformation

```python
for i in [1, 2, 3, 4, 5]:
    pyplot.hist((df['bullish'])**(1/i), bins = 40)
    pyplot.title("Bullish Transformation: 1/{}".format(str(i)))
    pyplot.show()
```

### Bearish Transformation

```python
for i in [1, 2, 3, 4, 5]:
    pyplot.hist((df['bearish'])**(1/i), bins = 40)
    pyplot.title("Bearish Transformation: 1/{}".format(str(i)))
    pyplot.show()
```

### User Generated Opinions Transformation

```python
for i in [1, 2, 3, 4, 5]:
    pyplot.hist((df['user_sent_opinions'])**(1/i), bins = 40)
    pyplot.title("User Generated Opinions Transformation: 1/{}".format(str(i)))
    pyplot.show()
```
